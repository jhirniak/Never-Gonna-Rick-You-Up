```python
# get_note = lambda t: (((t<<1)^((t<<1)+(t>>7)&t>>12))|t>>(4-(1^7&(t>>19)))|t>>7) % 256
get_rick = lambda t: int(((t<<3)*[8/9,1,9/8,6/5,4/3,3/2,0][[0xd2d2c8,0xce4088,0xca32c8,0x8e4009][t>>14&3]>>(0x3dbe4688>>(18 if (t>>10&15)>9 else t>>10&15)*3&7)*3&7]) % 256)
rick_notes = bytes([get_rick(t) for t in range(75200)][9820:])
```

```python
%matplotlib inline
import pandas as pd
pd.Series([get_rick(t) for t in range(75200)][9820:]).plot(figsize=(20,14))
```

```python
with open('rick_notes.txt', 'wb') as f:
    f.write(rick_notes)
```
